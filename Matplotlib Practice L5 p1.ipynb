# Importing The Matplotlib & numpy 

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Ex 1: Simple Line Plot 

```python
subjects = ['Math','Physic','Chemistry','English','CS']
marks = [78,85,72,80,90]
plt.plot(subjects,marks)
plt.title("marks VS Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

# Ex 2: Bar Chart 

```python
months = ['Jan','Feb','mar','Apr','May','Jun']
sales = [120,150,100,170,160,180]

plt.bar(months,sales)
plt.title("Monthly Sales ")
plt.show()
```

# Ex 3: Scatter Plot 

```python
study_hours = [1,2,3,4,5,6,7]
scores = [40,45,55,65,70,78,85]

plt.scatter(study_hours,scores)
plt.xlabel('Study Hours') 
plt.ylabel('Scores') 
plt.grid(True) 
plt.show()
```

# Ex 4: Multiple Lines 

```python
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'] 
city1 = [30,32,31,33,34,35,36] 
city2 = [25,26,27,26,28,29,30] 

plt.plot(days,city1,label='city A')
plt.plot(days,city2,label='City B')
plt.legend()
```

# Ex 5: Subplots 

```python
plt.subplot(2,1,1,)
plt.plot(subjects,marks)
plt.subplot(2,1,2)
plt.bar(months,sales)
plt.show()
```

# Ex 6: Pie Chart

```python
obtained = [78,85,72,80,90]
plt.pie(obtained,labels=subjects,autopct='%1.1f%%')
plt.title('Marks Distribution')
plt.show()
```

# Ex 7: Histogram 

```python
marks = np.random.randint(40,100,50) 
plt.hist(marks, bins=10) 
plt.title('Marks Histogram') 
plt.show()
```

# Ex 8: Box Plot

```python
marks = [0,1,2,3,4,5,6]
plt.boxplot(marks)
plt.title('Marks Boxplot')
plt.show()
np.random.seed(2) 
x=np.random.normal(5,5,20) # (5,5,200) 
plt.boxplot(x)
```

# Ex 9: Horizontal Bar Chart 

```python
countries = ['India','USA','UK','China']
population = [1400,330,67,1440]
plt.barh(countries , population)
plt.show()
```

# Ex 10 : Area Plot 

```python
usage = [120,130,125,140,150,160]
plt.fill_between(months,usage)
plt.title('Electricity Usage')
plt.show()
```

# Ex 11:Step Plot

```python
plt.step(study_hours, scores) 
plt.show() 
```

# Ex 12: Stem Plot 

```python
plt.stem(study_hours,scores)
plt.show()
```

# Ex 13: Log Scale Plot 

```python
plt.plot(study_hours,scores)
plt.yscale('log')
plt.show()
```

# Ex 14: Scatter with Size 

```python
sizes = [20,40,60,80,100,120,140]
plt.scatter(study_hours , scores , s = sizes)
plt.show()
```

# Ex 15: Annotated Plot 

```python
plt.plot(study_hours,scores)
plt.annotate('Top Score',(7,85))
plt.show()
```

# Ex 16: Grid Customization 

```python
subjects = [1,2,3,4,5]
marks = [78,85,72,80,90]
plt.plot(subjects, marks) 
plt.grid(True) 
plt.show()
```

# Ex 17: Line Styles and Markers 

```python
plt.plot(subjects, marks, linestyle='--', marker='o') 
plt.show() 
```

# Ex 18: 2x2 Subplots

```python
plt.subplot(2,2,1); plt.plot(subjects, marks) 
plt.subplot(2,2,2); plt.bar(months, sales) 
plt.subplot(2,2,3); plt.scatter(study_hours, scores) 
plt.subplot(2,2,4); plt.hist(marks) 
plt.show() 
```

# Ex 19: CSV Data Plot

# Ex: 22:

```python
products = ["Laptops", "Phones", "Tablets"]  # Categories 
sales = [200, 350, 150]                      
# Values 
plt.bar(products, sales, color=["#4CAF50", "#2196F3", "#FF5722"], width=0.6)                                
plt.title("Q1 Product Sales") 
plt.ylabel("Units Sold (Thousands)") 
plt.ylim(0, 400)   
# to add text on each  
for i, v in enumerate(sales): 
    plt.text(i, v, str(v), ha='center')
```
